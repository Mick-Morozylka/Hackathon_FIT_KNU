```python
import numpy as np
import pandas as pd
import csv
import psycopg2
```

```python
# підʼєднання до БД PostgreSQL
conn = psycopg2.connect("host=localhost dbname=mydatabase user=myuser password=mypassword")
cur = conn.cursor()

# Відкриття та запис файлу в БД
with open('data.csv', 'r') as f:
    reader = csv.reader(f)

    # Пропуск строки з заголовками
    next(reader)

    # Цикл по строках
    for row in reader:
        # запис даних 
        cur.execute("INSERT INTO mytable (column1, column2, column3) VALUES (%s, %s, %s)", row)

# Збереження змін та завершення процесу
conn.commit()
cur.close()
conn.close()
```

```python
piz = pd.read_csv('/Users/mishamac-pro/Downloads/pizza_sales/pizzas.csv')
order = pd.read_csv('/Users/mishamac-pro/Downloads/pizza_sales/orders.csv')
piztype = pd.read_csv('/Users/mishamac-pro/Downloads/pizza_sales/pizza_types.csv', encoding='latin1')
ordetail = pd.read_csv('/Users/mishamac-pro/Downloads/pizza_sales/order_details.csv')
```

```python
pizzas = pd.merge(piz, piztype, how='left', on='pizza_type_id')
```

```python
pizzas
```

```python
orders = pd.merge(order, ordetail, how='inner', on='order_id')
```

```python
orders
```

```python
pizza_df = pd.merge(orders, pizzas, how='left', on='pizza_id')
```

```python
pizza_df = pizza_df.drop(['pizza_id', 'pizza_type_id', 'order_details_id'], axis=1)
```

```python
pizza_df
```

```python
pizza_df['datetime'] = pd.to_datetime(pizza_df['date'] + ' ' + pizza_df['time'])
```

```python
df = pizza_df
```

```python
peak_hours = df.groupby(df['datetime'].dt.hour)[['quantity', 'price']].sum().sort_values('quantity', ascending=False)
```

```python
peak_hours 
#бачимо, що пікові години - це обід (з 12 до 13), а також район вечері (з 16 до 19).
```

```python
sales_by_month = df.set_index('datetime').resample('M')[['quantity', 'price']].sum().sort_values('quantity', 
                                                                                                 ascending=False)
```

```python
sales_by_month
#тут бачимо, що найкращі місяці згідно з продажами - це саме липень, травень і листопад.
```
